```
import pandas as pd
```

```
df_zone = pd.read_csv('nyc_locationid-longlat.csv')
df = pd.read_csv('YellowTaxi2016.csv')
```

```
df_zone.info()
```

```
df
```

```
for index, row in df.iterrows():
  for index1, row1 in df_zone.iterrows():
    if row['pulocationid'] == row1['location_i']:
      df.at[index, 'pu_lng'] = row1['lng']
      df.at[index, 'pu_lat'] = row1['lat']
  print(index, row['pulocationid'], row['pu_lng'], row['pu_lat'])
df
```

```
#Convert to FIFS Code
import requests
#Start_Point
df['puCensus_Tract'] = ''
for index, row in df.iterrows():
  if str(row['pu_lng']) != 'nan' and str(row['pu_lat']) != 'nan' and str(row['pu_lng']) != 0 and str(row['pu_lat']) != 0:
    url = 'https://geo.fcc.gov/api/census/block/find?latitude=' + str(row['pu_lat']) +'&longitude=' +str(row['pu_lng'])+'&showall=true&format=json'
    response = requests.get(url)
    resp = response.json()
    if 'Block' in resp:
      item = resp['Block']
      if len(item)>0:
        FIPS = item.get('FIPS')
        if FIPS != None:
          census_tract = FIPS[5:11].lstrip('0')
          print (index, census_tract)
          df.at[index,'puCensus_Tract'] = census_tract
df
```

```
df.to_csv('YellowTaxi2016CT.csv')
```
